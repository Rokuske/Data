```python
Train = df_Man1[df_Man1.Year < 2016].drop(['Unnamed: 0','ID', 'Name', 'Team', 'NOC', 'Games', 'Season', 'City', 'Sport', 'Event', 'Sex', 'Year'], axis=1)
Train
```

```python
Train['Medal'].value_counts()
```

```python
Test = df_Man1[df_Man1.Year >= 2016].drop(['Unnamed: 0','ID', 'Name', 'Team', 'NOC', 'Games', 'Season', 'City', 'Sport', 'Event', 'Sex', 'Year'], axis=1)
Test
```

```python
X_train = Train.drop(['Medal'], axis = 1)
y_train = Train.drop(['Age', 'Height', 'Weight'], axis = 1)
X_test = Test.drop(['Medal'], axis = 1)
y_test = Test.drop(['Age', 'Height', 'Weight'], axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
Random_Forest = RandomForestClassifier(n_estimators=100)
Random_Forest = Random_Forest.fit(X_train, y_train)
```

```python
y_pred = Random_Forest.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred, normalize=True)
```

```python
from sklearn.linear_model import LogisticRegression
LogisticReg = LogisticRegression(random_state=0).fit(X_train, y_train)
LogRegPred = LogisticReg.predict(X_test)
LogRegPred
```

```python
accuracy_score(y_test, LogRegPred, normalize=True)
```
