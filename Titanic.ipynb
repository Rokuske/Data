# 2.Работа с моделями

## 2.1.Данные для моделей

```python
Titanic_data
```

```python
Test_data
```

```python
k = 0.4 # задаём долю тестовых данных в нашем датасете 
mask = (np.random.random(Titanic_data.shape[0]) < k) # формируем маску для тестовых данных
X_train = Titanic_data[~mask][Titanic_data.columns.drop(['Survived', 'PassengerId'])]
# X - признаки, на основе которых модель будет делать предсказание
y_train = Titanic_data[~mask]['Survived']
X_train1 = Titanic_data.drop(columns = Titanic_data[['Survived', 'PassengerId']])
y_train1 = Titanic_data['Survived']
# y - целевой признак

X_test =Titanic_data[mask][Titanic_data.columns.drop(['Survived', 'PassengerId'])]

y_test = Titanic_data[mask]['Survived']
```

```python
X_train1
```

## 2.2.RandomForest

score 0.72727
n_estimators=20

score 0.73444

```python
from sklearn.ensemble import RandomForestClassifier
Random_Forest = RandomForestClassifier(n_estimators=100)
Random_Forest = Random_Forest.fit(X_train, y_train)
```

```python
y_pred = Random_Forest.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred, normalize=True)
```

```python
y_pred_test = Random_Forest.predict(Test_data.drop(columns = 'PassengerId'))
y_pred_test
```

```python
Test1 = pd.DataFrame()
Test1['PassengerId'] = Test_data['PassengerId']
Test1['Survived'] = y_pred_test
Test1 = Test1.set_index('PassengerId')
Test1
```

```python
Test1.to_csv('/content/gdrive/My Drive/Colab Notebooks/ML Studies/Titanic/predict.csv')
```
